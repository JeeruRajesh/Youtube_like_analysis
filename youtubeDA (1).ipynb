```python

c=widgets.Dropdown(
  options=video['Genre'].unique(),
  value='Music',
  description='Genre:',
  disabled=False,
)

def graph(a):
  dr = video[video['Genre']==a]
  gt = dr.groupby('channelTitle')
  gt=gt.sum().sort_values(by='view_count',ascending=False).head()
  sns.barplot(y=gt.index,x=gt['view_count'])
  plt.title(a)
  plt.xlabel('Views')
  plt.ylabel('Channel Name')
  plt.show()
```

```python
widgets.interact(graph,a=c)
```

```python

g=widgets.Dropdown(
  options=video['channelTitle'].unique(),
  value='NBA',
  description='channel:',
  disabled=False,
)

def graph2(z):
  ht = video[video['channelTitle']==z]
  ht=ht.sort_values(by='view_count',ascending=False).head()
  sns.barplot(y='title',x='view_count',data=ht)
  plt.title(z)
  plt.xlabel('views')
  plt.ylabel('video_name')
  plt.show()
```

```python
widgets.interact(graph2,z=g)
```

```python
k1=widgets.Dropdown(
  options=video['Genre'].unique(),
  value='Music',
  description='Genre:',
  disabled=False,)
def genre(m):
  gen=video[video['Genre']==m]
  k2=widgets.Dropdown(
  options=gen['channelTitle'].unique(),
  value=gen['channelTitle'].unique()[0],
  description='channel:',
  disabled=False,)
  def channel(d):
      ht=gen.sort_values(by='view_count',ascending=False).head()
      sns.barplot(y='title',x='view_count',data=ht)
      plt.title(d)
      plt.xlabel('views')
      plt.ylabel('video_name')
      plt.show()

  widgets.interact(channel,d=k2)
```

```python
 widgets.interact(genre,m=k1)
```
